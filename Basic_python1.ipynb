```
# Code Cell
# กดปุ่ม SHIFT + ENTER เพื่อรันโค้ดใน cell นี้
print("Hello World")

print("Anything")
```

```
# Basic Calculation
print(1 + 1)
print(2 * 3)
```

# **Text Cell**
ส่วนตรงนี้เราเรียกว่า TEXT CELL (หรือที่เรียกกันว่า Markdown Language)

เราสามารถเอาเมาส์ double click ที่ text cell เพื่อแก้ไขข้อความใน cell นี้ได้เลย ปกติเราใช้ text cell เพื่อจดโน้ตต่างๆ หรือใช้เขียนอธิบายโค้ดของเราให้คนอื่นที่มาอ่านโค้ดจะได้เข้าใจง่ายๆ

ลองกด SHIFT + ENTER รันโค้ดใน cell ถัดไป

```
# Example code
print("สวัสดีนักศึกษา !!")

while True:
  your_answer = input("\n2 * 7 มีค่าเท่ากับเท่าไหร่? ")
  if int(your_answer) == 14:
    print("\n14! ถูกต้องจ้าาาาา")
    break
  else:
    print("\nยังไม่ใช่จ้าาาา ลองตอบใหม่อีกทีซิ")
```
