## IF ELSE

if-else ใช้สำหรับทดสอบเงื่อนไข ถ้าเงื่อนไขเป็นจริง (True) ให้โปรแกรมทำอย่างหนึ่ง ถ้าเงื่อนไขไม่เป็นจริง (False) ก็ให้โปรแกรมทำอีกอย่าง

template สำหรับเขียน if-else ใน Python

---




```
if True:
    do something
else:
    do something else

---


```

Python coding style is very **elegant** เมื่อเทียบกับภาษาอื่นๆ คือมันไม่ได้ใช้ปีกกา { } ในการเขียน code block ให้ยุ่งยาก แต่จะใช้การย่อหน้าแทน (**indentation หรือ white space**)


---



## ตรงนี้สำคัญมาก !! 
ย่อหน้า (หรือ white space) ใน Python มีความหมาย โค้ดที่อยู่ในระนาบการย่อหน้าเดียวกัน Python จะเข้าใจว่ามันเป็น block เดียวกัน

มาลองดูตัวอย่างง่ายๆใน cell ด้านล่าง

````
# Example 1
# ก่อนจะรันโค้ดใน cell นี้
# ลองอ่านโค้ดดูก่อน คิดว่ามันจะปริ้นอะไรออกมาเป็น output? 

income = 20000

# simple if-else
if income <= 20000:
  print("You're quite poor!")
else:
  print("You're quite rich!")
````

````
# Example 2
# ก่อนจะรันโค้ดใน cell นี้
# ลองอ่านโค้ดดูก่อน คิดว่ามันจะปริ้นอะไรออกมาเป็น output? 

weather = "rainy"

if weather == "sunny":
  print("Good weather")
else:
  print("Bad weather because it's raining!")
````

## IF ELIF ELSE ทดสอบได้หลายๆเงื่อนไข

เราสามารถใช้ **if, elif, else** keyword ในการสร้างเงื่อนไขที่หลากหลายขึ้นได้

```
if condition1:
    do something
elif condition2:
    do something
elif condition3:
    do something
else:
    do something else
```



````
# Example 3 -- school grading scheme
# นักศึกษาสอบได้คะแนน 0-100
# มาลองตัดเกรดกัน

student_score = 75

# นักศึกษาคนนี้ จะได้เกรดอะไร?
````

````
if student_score >= 90:
  print("A")
elif student_score >= 80:
  print("B")
elif student_score >= 70:
  print("C")
elif student_score >= 60:
  print("D")
elif student_score >= 50:
  print("E")
else:
  print("Failed")
````

ลองเปลี่ยน student_score และทดลองรันโปรแกรมของเราใหม่ ว่าเกรดเปลี่ยนไหม?

ถ้า student_score = 44 เขาจะได้เกรดอะไร?

























## ฝึกฝน

ลองทำโจทย์ข้อนี้ด้วยตัวเอง 

มีเมนูอาหารที่เราชอบทาน 4 เมนู
-  weekday breakfast -- **bread and milk**
-  weekday dinner -- **boiled rice**
-  weekend breakfast -- **spaghetti**
-  weekend dinner -- **sukiyaki**

ถ้าวันนี้เป็นวันธรรมดา มื้อเย็น เมนูอาหารของเราคืออะไร?

```
## Hint
today = "weekday"
time = "dinner"

## ใช้ and เข้ามาช่วยสร้างเงื่อนไขได้
if (today == "weekday") and (time == "dinner"):
    print("boiled rice")
elif ...:
    ...
elif ...:
    ...
else:
    ...
```



````
# เขียนโค้ดนี้ให้เต็ม แล้วลองรันเพื่อเช็คคำตอบ
# ลองเปลี่ยน today, time ดูว่าได้คำตอบตรงอย่างที่เราต้องการไหม

today = "weekday"
time = "dinner"

if (today == "weekday") and (time == "dinner"):
  print("My favourite menu: " + "boiled rice")
elif (today == "weekday") and (time == "breakfast"):
  print("_____")
elif (today == "weekend") and (time == "dinner"):
  print("_____")
else:
  print("_____")
````
