## FOR LOOP

วิธีการเขียน for-loop คือบอกให้ Python ทำอะไรซ้ำๆไปเรื่อยๆ จนกว่าจะครบ loop นั้นๆ ตัวอย่างง่ายๆเช่น print("hello python") 10 รอบ เป็นต้น

```
text = "Hello Python."

for char in text:
    print(char)
```

สำหรับ char ที่อยู่ใน **for char in text:** เป็นแค่ชื่อ dummy จะเปลี่ยนเป็นอย่างอื่นก็ได้ เช่น

```
for i in text:
    print(i)
```

for loop ที่เราเขียนทั้งสองแบบได้ผลเหมือนกัน ลองรันโค้ดใน cell ด้านล่าง

````
# มาลองดู output ตรงนี้
text = "Hello Python."

# ปริ้น character ใน variable text
for i in text:
  print(i)
````

````
x = "Data Report"

for char in x:
  print(char)
````

## range() ฟังชั่นที่ใช้บ่อยใน Python

ฟัง์ชันใช้คู่กับ for-loop บ่อยๆ คือฟังชั่น range()

```
range(0, 10) จะสร้างตัวเลข 0-9 ขึ้นมา (10 not included)
range(6, 15) จะสร้างตัวเลข 6-14 ขึ้นมา (15 not included)
```
ลองรันโค้ดใน cell ด้านล่าง และดู output ที่ได้

````
for i in range(0,10):
  print(i)
````

````
for i in range(6, 15):
  print(i)
````

## range() มี 3 arguments (ค่า) ที่เราสามารถปรับได้

```
## explain range function
range(start, stop, step)

start = ตัวเลขแรกที่เริ่ม
stop = ตัวเลขสุดท้ายที่จบ - 1
step = คล้ายเป็น incremental จาก start ไปถึง stop

## example code
for i in range(1, 10, 2):
    print(i)
    
## output
1
3
5
7
9
```



````
for i in range(1, 20, 3):
  print(i)
````

````
# สั่ง Python ให้ print 'hello python' ห้ารอบ

for i in range(0, 5, 2):
  print("hello python")
````

## ลองเขียน for + ifelse เพื่อสร้างโปรแกรมที่ซับซ้อนขึ้น

เขียนโปรแกรมแสดงเลขคู่ระหว่าง (1, 21)

````
# print เฉพาะเลขคู่ (even number) เท่านั้น
# เลขคู่หารสองลงตัว แปลว่า i % 2 == 0

for i in range(1,21):
  if i % 2 == 0:
    print(i)
  else:
    continue # skip
````

````
# ลอง print เลขคี่ (odd number) 
for i in range(1,21):
  if i % 2 == 1: # i % 2 == 1 แปลว่าหารไม่ลงตัว เหลือเศษ 1
    print(i)
  else:
    continue # skip
````

## Summary for Loop


```
for i in range(1,21):
    if i % 2 == 0:
        print(i)
    else:
        continue
```

สิ่งที่ Python ทำใน for-loop นี้ คือ
1.  Python เอาตัวเลข 1 ไปจนถึง 20 ไปหารสอง (ทำทีละตัว)
2.  ถ้าเป็นเลขคู่ (i % 2 == 0) ปริ้นมันออกมา
3.  ถ้าเป็นเลขคี่ (i % 2 == 1) ข้ามไปเช็คตัวเลขตัวต่อไป ด้วย keyword **continue**

## WHILE LOOP


วิธีการเขียน while มีรูปแบบ คือ

```
while True:
    do something
```

````
# Example 1
x = 0

# print x ไปเรื่อยๆ ถ้า x <= 7
# ทุกครั้งที่ print(x) ให้เราเพิ่ม +1 ค่า x ด้วย

while x <= 5:
  print(x)
  x += 1
  # x = x+1
````

````
# Example 2
# หาค่า sqrt ของตัวเลขหนึ่งๆ

number = 25

while number/2 >= 1:
  print(number)
  number = number/2 # update number ใหม่
````

````
# Example 3
# สร้าง game ทายเลขง่ายๆ
# ถ้าทายถูก เลข 33 คือจบ
# แต่ถ้ายังทายไม่ถูกก้ต้องทายไปเรื่อยๆ

while True:
  guess_number = int(input("ทายเลข 1-100: "))
  if guess_number == 33:
    print("your guess is correct: 33")
    break # ทายถูกเราก็ break ออกจาก while-loop
  else:
    print("Try again")
    
````

อธิบาย ฟังชั่น **input()**  คือใช้รับ input จาก user

ค่าที่รับมาจะเป็น string เราเลยต้องใช้ **int(input())** เพื่อเปลี่ยนตัวเลขให้กลายเป็น int ก่อนจะไปเปรียบเทียบ if guess_number == 33: ใน if-else

## Summary while-loop

สำหรับ line สุดท้ายใน while-loop สองอันที่เราเขียนไป **สำคัญมาก** เราต้องอัพเดทค่าตัวแปรที่เราใช้เป็น condition ใน while loop 

ถ้าเราลืมอัพเดท while-loop จะวิ่งไปเรื่อยๆไม่หยุดเลย (infinite loop) โปรแกรมเราอาจค้างได้ 
```
x = 0
while x <= 10:
    do something
    x = x + 1 #อย่าลืมอัพเดทค่า x !! 
```
