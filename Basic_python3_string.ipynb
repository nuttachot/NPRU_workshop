## Slicing 
เราสามารถดึงข้อมูลออกจากตัวแปร course_name ได้ง่ายๆ Python เรียกวิธีการดึงค่าออกจากตัวแปรนึงๆว่า **Slice**

)

index ใน Python จะเริ่มจาก 0 และเราใช้เครื่องหมาย [  ] ในการ Slice ค่าที่ต้องการ

```
x = "data science"

ถ้าอยาก slice "d" ออกมาจากตัวแปร x พิมพ์ x[0]
ถ้าอยาก slice "a" ตัวที่สองออกมาจากตัวแปร x พิมพ์ x[1]

ถ้าอยาก slice "data" ออกมาจากตัวแปร x พิมพ์ x[0:4]
ถ้าอยาก slice "sci" ออกมาจากจากตัวแปร x พิมพ์ x[5:8]
ถ้าอยาก slice "science" ออกมาจากตัวแปร x พิมพ์ x[5:]

ถ้าอยาก slice "e" ตัวสุดท้ายออกมาจากตัวแปร x พิมพ์ x[-1]

เวลาเราเขียน [0:3] --> Python จะ slice index 0 ไปจนถึง 2 (แต่ไม่รวม 3)
เวลาเราเขียน [2:5] --> Python จะ slice index 2 ไปจนถึง 4 (แต่ไม่รวม 5)
```

**มาลอง slice ตัวแปร course_name **


````
print(course_name)
````

````
# อยากได้ตัว p ตัวแรกที่ index 0
course_name[0]
````

````
# อยากได้ตัว y ตัวที่สอง index 1
course_name[1]
````

````
# อยากไตัว r ตัวสุดท้าย index -1
course_name[-1]
````

````
# อยากได้ตัว e ตัวรองสุดท้าย index -2
course_name[-2]
````

````
# อยากได้คำว่า python
# python slice ตั้งแต่ index 0 ไปจนถึง 5 (แต่ไม่รวม 6)
# ภาษาอังกฤษ เราเรียกว่า up to but not include 6 (exclusive)
course_name[0:6]
````

````
# อยากได้คำว่า for
course_name[7:10]
````

````
# อยากได้คำว่า Data
course_name[11:15]
````

````
# อยากได้คำว่า Report
# slice ตั้งแต่ index 15 ไปจนสุดท้าย
course_name[16:]
````

## เราสามารถรวม string หลายๆคำเข้าด้วยกันด้วย +

ใช้เครื่องหมาย + ในการรวมคำได้เลย
```
greeting = "Hello"
name = "Nutt,"
question = "How are you?"

statement = greeting + name + question
print(statement)

```



````
greeting = "Hello"
name = "Nutt,"
question = "How are you?"

statement = greeting + name + question
print(statement)
````

````
# เวลาเราเคาะ spacebar เพื่อเว้นช่องว่าง
# ช่องว่างพวกนั้น ก็ถือว่าเป็น string character เหมือนกัน
# ลองพิมพ์ statement ใหม่อีกที

statement = greeting + " " + name + " " + question
print(statement)
````

dkii;, string เข้าด้วยกัน ต้อง แน่ใจว่าทุกตัวแปรที่เอามาบวกกัน ต้องเป็น string data type

เราสามารถใช้ฟังชั่น **str()** ในการเปลี่ยน type ของตัวแปรก่อนมารวมกันได้ง่ายๆ

````
# สร้างตัวแปรส่วนสูง น้ำหนัก อายุ แบบ int 
height = 161
weight = 61
age = 32

print(type(age))
````

````
# เปลี่ยนตัวแปร height, weight, age ให้เป็น string ก่อนจะปริ้นใน console
# ทบทวน \n คือการขึ้นบรรทัดใหม่ (new line)
print("my height: " + str(height) + "\nmy weight: " + str(weight) + "\nmy age: " + str(age))
````

````
# ถ้าเราลืมเปลี่ยนจาก int --> str เวลาสั่ง print() มันจะ error
print("my height: " + height) # ขึ้น error แก้ง่ายด้วยการเปลี่ยน height เป็น str(height)
````

## Optional: Another Cool Trick for Printing String

เราสามารถใช้ .format( ) เวลาเรา print string ลองดูตัวอย่างการใช้งานด้านล่าง

```
string.format( )
```

เราแทนที่ตัวแปรของเราลงไปใน placeholder { } ของ string ที่เราต้องการได้เลย

````
height = 161
weight = 61
age = 32

print("Hi, my name is PooH. Height: {}, Weight: {}, Age: {}".format(height, weight, age))
````
